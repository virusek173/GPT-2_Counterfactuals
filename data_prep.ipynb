```python
import pandas as pd
from tqdm import tqdm_notebook

prefix = 'data/'
```

```python
file_patch = 'gpt2_generated_final.csv'
train_df = pd.read_csv(prefix + file_patch, header=None)
train_df.head()
```

```python
file_patch = 'gpt2_generated_final.csv'
test_df = pd.read_csv(prefix + file_patch, header=None)
test_df.head()
```

```python
train_df.head()
```

```python
train_df = pd.DataFrame({
    'id':range(len(train_df)),
    'label':train_df[1],
    'alpha':['a']*train_df.shape[0],
    'text': train_df[2].replace(r'\n', ' ', regex=True)
})

train_df.head()
```

```python
dev_df = pd.DataFrame({
    'id':range(len(test_df)),
    'label':test_df[1],
    'alpha':['a']*test_df.shape[0],
    'text': test_df[2].replace(r'\n', ' ', regex=True)
})

dev_df.head()
```

```python
train_df.to_csv('data/train.tsv', sep='\t', index=False, header=False, columns=train_df.columns)
dev_df.to_csv('data/dev.tsv', sep='\t', index=False, header=False, columns=dev_df.columns)
```
